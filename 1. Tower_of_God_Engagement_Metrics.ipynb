# Visualization of Engagement Metrics for "Why is the Tower of God Show So Popular ?"

```python
import matplotlib.pyplot as plt
```

```python
page_views = 10000
avg_time_spent = 2
bounce_rate = 75
```

```python
fig, ax = plt.subplots()
metrics = ['Page Views', 'Avg Time Spent (min)']
values = [page_views, avg_time_spent]

ax.bar(metrics, values, color=['skyblue', 'salmon'])

ax.set_title('Engagement Metrics for "Why is the Tower of God Show So Popular?"')
ax.set_ylabel('Value')
plt.show()
```

```python
# Pie chart for Bounce rate
fig, ax = plt.subplots()

labels = ['Bounce', 'Stay']
sizes = [bounce_rate, 100 - bounce_rate]
colors = ['lightcoral', 'yellowgreen']

ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
ax.axis('Equal')

plt.title('Bounce Rate Analysis')
plt.show()
```
